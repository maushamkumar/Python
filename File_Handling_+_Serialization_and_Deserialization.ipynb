# Using Context Manager (With)
* It's a good idea to close a file after usage as it will free up the resources.
* If we don't close it, the garbage collector will close it
* With keyword closes the file as soon as the usage is over.

```python
# with
with open('sample.txt', 'w') as f:
  f.write("Mera bhai")
```

```python
# Try f.read() now
with open('sample.txt', 'r') as f:
  print(f.read())
```

```python
# Moving within a file -> 10 char then 10 char
with open('sample.txt', 'r') as f:
  print(f.read(10))
  print(f.read(10))
```

```python
# Benefit -> to load a big file in memory
big_l = ['Hello Mausham\n' for i in range(10000)]
with open('big.txt', 'w') as f:
  f.writelines(big_l)
```

```python
with open('big.txt', 'r') as f:
  print(f.read(100))
```

```python
with open('big.txt', 'r') as f:
  chunk_size = 100
  while len(f.read(chunk_size)) > 0:
    print(f.read(chunk_size))
    f.read(chunk_size)
```

```python
# Seek and tell function
with open('sample.txt', 'r') as f:
  print(f.read(10))
  print(f.tell()) # tell tells the position of the cursor
  f.seek(4)  # seek() is used to move the cursor
  print(f.read(10))
```

# Serialization and Deserialization
* **Serialization** - Process of converting python data types to json
* **Deserialization** - Precess of converting JSON to python data types
What is **JSON**

```python
# Serialization using json module
# list
import json
L = [1, 2, 3, 4]

with open('demo.json', 'w') as f:
  json.dump(L, f)
```

```python
# dict
d = {
    'name' : 'Mausham',
    'age' : 19,
    'gender' : 'Male'
}

with open('demo.json', 'w') as f:
  json.dump(d, f, indent = 4)
```

```python
# Deserialization
import json

with open ('demo.json', 'r') as f:
  d = (json.load(f))
  print(d)
  print(type(d))
```

```python
# Serialixe and deserialize tuple
import json
t = (1, 2, 3, 4)
with open('demo.json', 'w') as f:
  json.dump(t, f)
```

```python
# Serialize and deserialize a nested dict
d = {
    'student' : 'Mausham',
    'Marks' : [23, 2, 34, 45, 48]

}

with open('demo.json', 'w') as f:
  json.dump(d, f)
```

# Serializing and Deserializing custom objects

```python
class Person:
  def __init__(self, fname, lname, age, gender):
    self.fname = fname
    self.lname = lname
    self.age = age
    self.gender = gender

# format to printed in
# -> Mausham kumar age -> 19 -> gender -> Male
```

```python
person = Person('Mausham','Kumar', 19, 'Male')
```

```python
# As a string
import json
with open('demo.json', 'w') as f:
  json.dump(person, f)
```

```python
# As a string
import json

def show_object(person):
  if isinstance(person, Person): # WE are checking person person class ka object h ki nahi
    return "{} {} age -> gender -> {}".format(person.fname, person.lname, person.age, person.gender)

with open('demo.json', 'w') as f:
  json.dump(person, f, default=show_object)
```

```python
# As a dict
import json

def show_object(person):
  if isinstance(person, Person): # WE are checking person person class ka object h ki nahi
    return {'Name' : person.fname + " " + person.lname,
            'age' : person.age,
            'Gender' : person.gender}
with open('demo.json', 'w') as f:
  json.dump(person, f, default=show_object, indent = 4)
```

```python
# Deserializing
import json
with open('demo.json', 'r') as f:
  print(json.load(f))
```

## Pickling
Pickling is the process whereby a Python object hierarchy is converted into a byte stream, and Upickling is the inverse operation, whereby a byte stream (from a binary file or bytes-like object) is converted back into an object hierarchy.

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def display_info(self):
    print('Hi my name is', self.name, 'and I am', self.age, 'years old')
```

```python
p = Person('Mausham', 19)
```

```python
# Pickle dump
```

```python
import pickle
with open('person.pkl', 'wb') as f:
  pickle.dump(p,f)
```

```python
# Picklee load
with open('person.pkl', 'rb') as f:
  p = pickle.load(f)
p.display_info()
```
